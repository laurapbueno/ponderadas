```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exploração dos Dados

```python
btc_df = pd.read_csv('../data/raw/coin_Bitcoin.csv')
eos_df = pd.read_csv('../data/raw/coin_EOS.csv')
sol_df = pd.read_csv('../data/raw/coin_Solana.csv')

print(btc_df.head())
print(eos_df.head())
print(sol_df.head())
```

```python
# Verificado a estrutura dos dados
print(btc_df.info())
print(eos_df.info())
print(sol_df.info())

# Verificando estatísticas descritivas:
print(btc_df.describe())
print(eos_df.describe())
print(sol_df.describe())
```

```python
# Plotando os preços de fechamento ao longo do tempo para Bitcoin
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(btc_df['Date']), btc_df['Close'], label='Bitcoin')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Preço de Fechamento do Bitcoin ao Longo do Tempo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotando os preços de fechamento ao longo do tempo para EOS
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(eos_df['Date']), eos_df['Close'], label='EOS', color='orange')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Preço de Fechamento da EOS ao Longo do Tempo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotando os preços de fechamento ao longo do tempo para Solana
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(sol_df['Date']), sol_df['Close'], label='Solana', color='green')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Preço de Fechamento da Solana ao Longo do Tempo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotando o preço de fechamento das 3 moedas
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(btc_df['Date']), btc_df['Close'], label='Bitcoin', color='blue')
plt.plot(pd.to_datetime(eos_df['Date']), eos_df['Close'], label='EOS', color='orange')
plt.plot(pd.to_datetime(sol_df['Date']), sol_df['Close'], label='Solana', color='green')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Preço de Fechamento de Bitcoin, EOS e Solana ao Longo do Tempo')
plt.legend()
plt.grid(True)
plt.show()
```

# Pré-processamento

```python
print(btc_df.isnull().sum())
print(eos_df.isnull().sum())
print(sol_df.isnull().sum())


# Tratando valores ausentes - se houver - (método: preencher com o valor anterior)
btc_df.fillna(method='ffill', inplace=True)
eos_df.fillna(method='ffill', inplace=True)
sol_df.fillna(method='ffill', inplace=True)
```

```python
# Convertendo a coluna de datas para o tipo datetime
btc_df['Date'] = pd.to_datetime(btc_df['Date'])
eos_df['Date'] = pd.to_datetime(eos_df['Date'])
sol_df['Date'] = pd.to_datetime(sol_df['Date'])
```

```python
# Criação de uma média móvel de 30 dias
btc_df['SMA30'] = btc_df['Close'].rolling(window=30).mean()
```

# EDA

```python
btc_numeric_df = btc_df.select_dtypes(include=[np.number])

plt.figure(figsize=(10, 7))
sns.heatmap(btc_numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação para Bitcoin')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.histplot(btc_df['Close'], bins=50, kde=True)
plt.title('Distribuição dos Preços de Fechamento do Bitcoin')
plt.show()
```
